```python
#filling the null values with appropriate data

for cols in df:
    if(df[cols].dtype == 'object' and df[cols].isna().any()):
        df[cols].fillna('Missing', inplace = True)
    elif(df[cols].dtype == 'float64' and df[cols].isna().any()):
        df[cols].fillna(0.0, inplace = True)
        
df.head()
```

```python
#checking if the value update was successful. {should be all false}
df.isna().any()
```

```python
#remove decimal from the numeric values
for i in df:
    if(df[i].dtype == 'float64'):
        df[i] = df[i].astype('int64')
```

```python
#display pairs to see which two columns have highest correlation
corr_matrix = df_num.corr()
corr_pairs = corr_matrix.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
highest_corr = sorted_pairs[(sorted_pairs)>0.5]
highest_corr
```
